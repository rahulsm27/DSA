```python
# 2 Print circular link list..seems easy but is slight TRICKY..check

class Node:
    def __init__(self, k):
        self.key=k
        self.next = None

def printCirLink(head):

    if head == None:
        return None
    
    print(head.key, end=" ") # print first node otherwise we will not be able to enter while loop
    curr = head.next
    while curr != head:
        print(curr.key, end=" ")
        curr = curr.next

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = head
printCirLink(head)
```

```python
# 3 Insert at the begining of circluar linked list
# this take n



def insertBegin(head,data):
    new_head = Node(data)

    if head == None:
        new_head.next = head
        return new_head

    curr = head
    while curr.next != head:
        curr = curr.next

    curr.next = new_head
    new_head.next = head

    return new_head
```
