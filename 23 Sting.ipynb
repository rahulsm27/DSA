```python
# 3 PATTERN SEARCHING 
#Time complexity O((n-m+1) * m) -> two loops one outer and one inner
```

![Alt text](image-232.png)

```python
def patSearch(string,patt):

    n = len(string)
    m = len(patt)

    for i in range(n-m+1):
        for j in range(m):
            if string[i+j] != patt[j]:
                break
        else:
            print(i,end=" ")
```

```python
test =  [('ABABABCD','ABAB'),('ABCABCD','ABCD'),('AAAAA','AAA')]
for s,p in test:
    print()
    print(s,p)
    patSearch(s,p)
```

```python
# 4 NAIVE APPROACH IN LINEAR TIME
```

```python
def patSearch(string,patt):

    n = len(string)
    m = len(patt)

    i = 0

    while i <= (n-m+1):
        j = 0
        while j < m:
            if string[i+j] != patt[j]:
                break
            else:
                j = j+1
        if j == m:
            print(i,end= " ")
        if j == 0:
            i = i+1
        else:
            i = i+j    
test =  [('ABCABABCD','ABC'),('GEEKSFORGEEKS','EKS')]
for s,p in test:
    print()
    print(s,p)
    patSearch(s,p)
```
