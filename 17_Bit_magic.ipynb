```python
# 3 Count the number of set bits.

# traver only through set bits.

# in each iteration make last set bit  == 0
# in this way we can traverse only thorugh number of set bits 
"""
 1  Initialize count: = 0
 2  If integer n is not zero
      (a) Do bitwise & with (n-1) and assign the value back to n
          n: = n&(n-1)
      (b) Increment count by 1
      (c) go to step 2
 3  Else return count

 """
```

![Alt text](image-54.png)

![Alt text](image-55.png)

```python
# Function to get no of set bits in binary
# representation of passed binary no. */
def countSetBits(n):

	count = 0
	while (n):
		n &= (n-1)
		count+= 1
	
	return count


# Program to test function countSetBits
i = 9
print(countSetBits(i))
```

```python
# 4 Power of 2

#   n & N-1 will give number with right most bit set to 0 of n

# n & -(n-1) will give the rightmost bit
```

```python
#Another way is to use the logic to find the rightmost bit set of a given number and then check if (n & (~(n-1))) is equal to n or not

def isPowerofTwo(n):

	if (n == 0):
		return 0
	if ((n & (~(n - 1))) == n):
		return 1
	return 0
```

```python
# Prints two numbers that occur odd
# number of times. The function assumes
# that the array size is at least 2 and
# there are exactly two numbers occurring
# odd number of times.
def printTwoOdd(arr, size):
	
	# Will hold XOR of two odd occurring elements
	xor2 = arr[0]
	
	# Will have only single set bit of xor2
	set_bit_no = 0
	n = size - 2
	x, y = 0, 0

	# Get the xor of all elements in arr[].
	# The xor will basically be xor of two
	# odd occurring elements
	for i in range(1, size):
		xor2 = xor2 ^ arr[i]

	# Get one set bit in the xor2. We get
	# rightmost set bit in the following
	# line as it is easy to get
	set_bit_no = xor2 & ~(xor2 - 1)

	# Now divide elements in two sets:
	# 1) The elements having the corresponding bit as 1.
	# 2) The elements having the corresponding bit as 0.
	for i in range(size):
	
		# XOR of first set is finally going to
		# hold one odd occurring number x
		if(arr[i] & set_bit_no):
			x = x ^ arr[i]

		# XOR of second set is finally going
		# to hold the other odd occurring number y
		else:
			y = y ^ arr[i]

	print("The two ODD elements are", x, "&", y)

# Driver Code
arr = [4, 2, 4, 5, 2, 3, 3, 1]
arr_size = len(arr)
printTwoOdd(arr, arr_size)
```
