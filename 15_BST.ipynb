```python
# 2 SEARCH IN BST

def search(root,data):

    if root == None:
        return False
    elif root.data == data :
        return True
    else:
        if root.data<data:
            return search(root.left)
        else:
            return search(root.right)
    
```

```python
def delete(root,key):
    if root == None:
        return

    elif root.data == key:
        if root.left == None and root.right == None:
            root.data = None
            return
        elif root.left!= None and root.right == None:
            root = root.left
            return
        elif root.left == None and root.right != None:
            root = root.right
            return
        else:
            curr = root.right
            while curr.left != None:
                curr = curr.left
            root.data = curr.data
            curr.data = None
            

    else:
        if root.data < key:
            delete(root.right)
        else:
            delete(root.left) 
            


```

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.key = key
        self.right = None



def getSucc(curr, key):
    while curr.left != None:
        curr = curr.left

    return curr.key


def deleteNode(root, key):
    if root == None:
        return

    if root.key > key:
        root.left = deleteNode(root.left, key)

    if root.key < key:
        root.right = deleteNode(root.right, key)

    else:
        if root.left == None:
            return root.right
        elif root.right == None:
            return root.left
        else:
            succ = getSucc(root.right, key)
            root.key = succ
            root.right = deleteNode(root.right, succ)

    return root
```

```python
# Binary Tree Node 
""" A utility function to create a
new BST node """
class newNode: 

    # Construct to create a newNode 
    def __init__(self, data): 
        self.data = data
        self.left = None
        self.right = None
""" This function is used to insert 
new values in BST"""
def insert(root, key): 

    if (not root):
        return newNode(key) 
    if (key < root.data): 
        root.left = insert(root.left, key) 
    else:
        root.right = insert(root.right, key) 
    return root 
```
