```python
def per(s,i,n):

    if i == n:
        print(s)
        return

    else:

        for j in range(i,n):
           # print(j)
            s[i] ,s[j] = s[j],s[i]
            per(s,i+1,n)
            s[i] ,s[j] = s[j],s[i]
        

per(['A','B','C'],0,3)

```
